# Data Visualization With Seaborn :
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

- Basic ploting with seaborn

```python
tips =sns.load_dataset('tips')
tips
```

- Create Scatter plot

```python
sns.scatterplot(x='total_bill',y='tip',data=tips,color='green')
plt.title('Scatterplot of total bill vs tip')
plt.show()
```

- Line plot

```python
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total bill by size')
plt.show()
```

- Categorical Plots


```python
# Bar plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of Total Bill by Day')
plt.show()
```

```python
# Box Plot : useful to check an outliers
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title('total bill for days')
plt.show()
```

```python
# Violin plot : to see the distribution of data
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin plot for total bill for particular day')
plt.show()
```

```python
# Histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

- Pair Plot 
  - to see the relation between diffrent features
  - applies to numerical values not for categorical

```python
sns.pairplot(tips)
```

Heat map : for corelation

```python
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

#### Cutomizing plot 

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# Ploting total sales by product
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales by Product')
plt.xlabel('Products')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Ploting total sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Regions')
plt.ylabel('Total Sales')
plt.show()
```
